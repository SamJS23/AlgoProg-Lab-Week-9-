```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def get_name(self):
        return self.name
    def get_age(self):
        return self.age
    def set_name(self,new_name):
        if isinstance(new_name,str):
            self.name = new_name
        else:
            print("Invalid input for the name. Please provide a string.")
    def set_age(self,new_age):
        if isinstance(new_age,int) and 0 <= new_age <= 150:
            self.age = new_age
        else:
            print("Invalid input for age")
#Abstraction 
    def display_info(self):
        print("Person's Information: ")
        print(f"    - Name: {self.name}")
        print(f"    - Age: {self.age}")
#Inheritence
class Student(person):
    def __init__(self,name,age, student_grade):
        super().__init__(name,age)
#Encapsulation
        self.__student_grade = student_grade
    def get_student_grade(self):
        return self.__student_grade
    def set_student_grade(self,new_grade):
        self.__student_grade = new_grade
    def display_info(self):
        print("Student Information: ")
        print(f"    - Name: {self.name}")
        print(f"    - Age: {self.age}")
        print(f"    - Grade: {self.__student_grade}")
class Working_Adult(person):
    def __init__(self,name,age,job):
       super().__init__(name,age) 
       self.__job = job
    def get_job(self):
        return self.__job
    def set_job(self,new_job):
        self.__job = new_job
#Polymorphism (overriding)
    def display_info(self):
        print("Working Adult's Information: ")
        print(f"    - Name: {self.name}")
        print(f"    - Age: {self.age}")
        print(f"    - Job: {self.__job}")
        
student1 = Student(name = "Samuel", age = 18, student_grade="12th grade")
x = student1.get_name()
print(f'Name: {x}')
x = student1.get_student_grade()
print(f'Grade: {x}')
student1.set_age(18)
student1.set_student_grade("11th grade")
student1.display_info()
print()

adult1 = Working_Adult(name="Samuel", age = 20, job ="Software Engineer")
x = adult1.get_name()
print(f'Name: {x}')
x = adult1.get_job()
print(f'Job: {x}')
adult1.set_age(24)
adult1.set_job("11th grade")
adult1.display_info()
```
